Aim: Describing the Distribution of a Variable:
a. Obtain a dataset and calculate descriptive statistics (mean, median, mode, variance, etc.) for a specific variable of interest.Use 'OnlineRtail.csv' dataset.
b. Create visualizations (histograms, box plots) to depict the distribution of a variable and analyze its characteristics.

```python
import pandas as pd
file_path = 'OnlineRetail.csv'
data = pd.read_csv(file_path,encoding='latin1')

data.head()
```

```python
mean=data[['Quantity','UnitPrice']].mean()
print(mean)
quantity_mode = data['Quantity'].mode()
print(f"Mode : {quantity_mode.values}")
```

```python
#Describe stastistices for quantity
quantity_stats = data['Quantity'].describe()
print("Descriptive Satiestics for Quantity")
print(quantity_stats)
```

```python
#Compute variance and standard deviation
quantity_variance =data['Quantity'].var()
quantity_std = data['Quantity'].std()

print(f"Variance: {quantity_variance}")
print(f"Standard Deviation: {quantity_std}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#Set an the figure size and style
plt.figure(figsize=(14, 6))
sns.set_style("whitegrid")

#Histogram for Quantity
plt.subplot(1, 2, 1)
sns.histplot(data['Quantity'], bins=30, kde=True, color='blue')
plt.title('Histogram of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')

# Box plot for Quantity
plt.subplot(1, 2, 2)
sns.boxplot(x=data['Quantity'], color='lightgreen')
plt.title('Box Plot of Quantity')
plt.xlabel('Quantity')
```

```python
#focus on a narrower range of UnitPrice (1st to 95th percentile)
lower_ound = data['UnitPrice'].quantile(0.01)
upper_bound = data['UnitPrice'].quantile(0.95)

filtered_data = data[(data['UnitPrice'] >= lower_ound) & (data['UnitPrice'] <= upper_bound)]

plt.figure(figsize=(14, 6))
sns.set_style("whitegrid")

# Box plot for UnitPrice
sns.boxplot(x=filtered_data['UnitPrice'], color='pink')
plt.title('Box Plot of Unit Price (lower)')
plt.xlabel('Unit Price')

plt.show()
```
